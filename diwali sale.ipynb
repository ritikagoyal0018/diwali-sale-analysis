## EXPLORATORY ANALYSIS

```python
df.columns
```

# which gender buys more

```python
color=['red','blue']
plt.figure(facecolor='lightgrey')
sns.set_style("dark")
ax=sns.countplot(x='Gender',data=df,palette=color)
for bars in ax.containers:
    ax.bar_label(bars)
    
```

# age group

```python
sns.countplot(x='Age Group',data=df,palette='Set2')
```

```python
color=["orange","yellow"]
ax=sns.countplot(x='Age Group',data=df,palette=color,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

# state

```python
df.columns
```

```python
sales = df.groupby(['Gender'])['Amount'].sum().head(10)
sales
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().head(10)
plt.figure(figsize=(15, 8))
sns.barplot(data = sales_state, x = 'State',y= 'Orders',palette="Set2")
```

```python
gen = df.groupby(['Gender'])['Amount'].sum().reset_index()
color=["orange","yellow"]

sns.barplot(data=gen, x='Gender', y='Amount',palette=color)
```

```python
df.columns
```

```python
sales_orders=df.groupby(['State'] ,as_index=False)['Orders'].sum().head(10)
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Orders',data=sales_orders,palette="Set2")
```

# Marital Status

```python
color=["orange","yellow"]
ax=sns.countplot(data=df,x='shaadi',palette=color,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python

sales_status=df.groupby(['shaadi','Gender'],as_index=False)['Amount'].sum().head(10)
sns.barplot(data = sales_status, x = 'shaadi',y= 'Amount',hue='Gender')
```

# Occupation

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(x='Occupation',data=df)
```

```python
occupation_status=df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum()
plt.figure(figsize=(20,5))
sns.barplot(data=occupation_status,x='Occupation',y='Amount',hue='Gender')
```

# product category

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product_status=df.groupby(['Product_Category','Gender'],as_index=False)['Amount'].sum()
plt.figure(figsize=(20,5))
sns.barplot(data=product_status,x='Product_Category',y='Amount',hue='Gender')
```
